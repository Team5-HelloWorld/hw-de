```python
chars = 'abcdefghijklmnopqrstuvwxyz'
hyphen_dash = '-_'
nums = '0123456789'
    
def random_email_naver(c_leng, dash, n_leng):
    c_result = ''
    n_result = ''
    hy_result = ''
    
    for i in range(c_leng):
        c_result += random.choice(chars)
    for i in range(dash):
        hy_result += random.choice(hyphen_dash)
    for i in range(n_leng):
        n_result += random.choice(nums)
        
    return c_result+hy_result+n_result+"@naver.com"
```
